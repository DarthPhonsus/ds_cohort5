```python
import pyspark.sql.functions as f
from pyspark.sql.types import FloatType
from pyspark.sql.types import DoubleType
@f.udf(returnType=FloatType())
def kms(coord1, coord2, coord3, coord4):
  return geopy.distance.distance((coord1, coord2), (coord3, coord4)).km
```

```python
df2 = (
  df
    .withColumn('merchant_distance', kms(
      f.col('lat'), 
      f.col('long'), 
      f.col('merch_lat'), 
      f.col('merch_long')
    ))
)
```

```python
from pyspark.sql.functions import current_date
from pyspark.sql.functions import datediff, to_date, lit
from pyspark.sql.functions import floor
from pyspark.sql.functions import dayofweek


test_df = df2.withColumn("current_date",current_date())
test_df = test_df.withColumn("dob", to_date(test_df.dob, 'yyyy-MM-dd'))
test_df = test_df.withColumn("trans_date", to_date(test_df.trans_date, 'yyyy-MM-dd'))
# for day_of_week, 1 = Sunday, 2 = Monday, 3 = Tuesday, ....
test_df = test_df.withColumn('day_of_week',dayofweek(test_df.trans_date))

test_df = test_df.withColumn("age", 
                            datediff(test_df.current_date, test_df.dob))
test_df = test_df.withColumn("age", floor(test_df.age / 365 ) )
test_df = test_df.withColumn("amt_floor", floor(test_df.amt) )
display(test_df.head(5) )
```

```python

columns_to_drop = ['ssn', 'first', 'last', 'gender', 'street', 'city', 'zip', 'lat', 'long', 'city_pop', 'amt_floor', 'current_date' ]



df_new = df3.drop(*columns_to_drop)

display(df_new.head(10))
```

```python
#lets convert all nums to double type

df_new = df_new.withColumn("is_fraud", df_new.is_fraud.cast("double"))
df_new = df_new.withColumn("day_of_week", df_new.day_of_week.cast("double"))
df_new = df_new.withColumn("age", df_new.age.cast("double"))
df_new = df_new.withColumn("hour", df_new.hour.cast("double"))
#converting cc_num to string for string indexer then OHE
df_new = df_new.withColumn("cc_num", df_new.cc_num.cast("string"))
```

```python
display(df_new.head(10))
```

```python
df_new.printSchema()
```

```python
string_df = df_new.select("cc_num", "state", "category", "merchant")
```

```python
small_test_df = df_new.sample(False, 0.01)
```

```python
display(small_test_df.head(10) )
```

```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.feature import OneHotEncoder, VectorAssembler
'''
encoder = OneHotEncoder(
    inputCols=[indexer.getOutputCol() for indexer in indexers],
    outputCols=[
        "{0}_encoded".format(indexer.getOutputCol()) for indexer in indexers]
)

assembler = VectorAssembler(
    inputCols=encoder.getOutputCols(),
    outputCol="features"
)
'''
indexer = []
for c in string_df.columns: 
  indexer.append(StringIndexer(inputCol=c, outputCol='{}_index'.format(c) ) )

encoder = OneHotEncoder(inputCols=["cc_num_index", "state_index", "category_index", "merchant_index" ], 
                        outputCols= ["cc_num_encoded", "state_encoded", "category_encoded", "merchant_encoded" ] )

for index in indexer:
  indexerModel = index.fit(small_test_df)
  small_test_df = indexerModel.transform(small_test_df)
```

```python
display(small_test_df.head(10) )
```

```python
small_test_df.printSchema()
```

```python
featureCols = small_test_df.columns
drop_list = ["is_fraud", "cc_num", "state", "job", "dob",
                  "acct_num", "trans_num", "trans_date", "trans_time",
                  "unix_time", "category", "merchant", "merchant_distance",
            ]

featureCols.remove("is_fraud")
featureCols.remove("cc_num")
featureCols.remove("state")
featureCols.remove("job")
featureCols.remove("dob")
featureCols.remove("acct_num")
featureCols.remove("trans_num")
featureCols.remove("trans_date")
featureCols.remove("trans_time")
featureCols.remove("unix_time")
featureCols.remove("category")
featureCols.remove("merchant")
featureCols.remove("merchant_distance")

```

```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=featureCols, outputCol="rawFeatures")

pipeline = Pipeline(stages=[encoder, assembler])
pipelineModel = pipeline.fit(small_test_df)

predictionsDF = pipelineModel.transform(small_test_df)
```
